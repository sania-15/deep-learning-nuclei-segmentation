# Renaming all images to [folder]_image.png

```python
import os
from tqdm import tqdm

# Base paths
he_base = "/content/drive/MyDrive/ProjectMIA/FinalFolder/3. HEtiles_small"
ca_base = "/content/drive/MyDrive/ProjectMIA/FinalFolder/2. CAtiles_small"

# Function to rename images in all subfolders
def rename_all_folders(base_path):
    folders = sorted([f for f in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, f))])
    print(f"\nüìÇ Found {len(folders)} folders in: {base_path}")

    for folder in folders:
        folder_path = os.path.join(base_path, folder)
        files = sorted([f for f in os.listdir(folder_path) if f.endswith(".png")])
        print(f"\nüîÅ Renaming {len(files)} files in: {folder_path}")
        for i, filename in enumerate(tqdm(files, desc=f"Renaming {folder}"), start=1):
            src = os.path.join(folder_path, filename)
            dst_name = f"{folder}_image{i}.png"
            dst = os.path.join(folder_path, dst_name)
            os.rename(src, dst)
        print(f"‚úÖ Done renaming files in: {folder_path}")

# Run for both HEtiles_small and CAtiles_small
rename_all_folders(he_base)
rename_all_folders(ca_base)

print("\n‚úÖ All images in all folders renamed.")
```

## cellpose segmentation

```python
import os
from glob import glob
from tqdm import tqdm
from skimage.io import imread
from cellpose.models import CellposeModel
from cellpose import io

# Load Cellpose model
model = CellposeModel(gpu=True)

# Define input and output directories
input_base = "/content/drive/MyDrive/ProjectMIA/FinalFolder/3. HEtiles_small"
output_base = "/content/cellpose_outputs"
os.makedirs(output_base, exist_ok=True)

# Get all 33 folders
folders = sorted([f for f in os.listdir(input_base) if os.path.isdir(os.path.join(input_base, f))])
print(f"üìÇ Found {len(folders)} folders.")

# Segment images folder by folder
for folder in tqdm(folders, desc="üì¶ Processing folders"):
    input_path = os.path.join(input_base, folder)
    output_path = os.path.join(output_base, folder)
    os.makedirs(output_path, exist_ok=True)

    image_paths = sorted(glob(os.path.join(input_path, "*.png")))
    for img_path in tqdm(image_paths, desc=f"üîç {folder}", leave=False):
        img = imread(img_path)

        # Determine channels
        if img.ndim == 2:
            channels = [0, 0]
        elif img.ndim == 3 and img.shape[-1] == 3:
            channels = [2, 1]
        else:
            print(f"‚ö†Ô∏è Skipping: {img_path} (unexpected shape {img.shape})")
            continue

        # Run Cellpose
        masks, _, _ = model.eval(img, diameter=None, channels=channels)

        # Save mask
        base = os.path.basename(img_path).replace(".png", "")
        io.save_masks(img, masks, None, base, savedir=output_path)

print("\n‚úÖ Cellpose segmentation completed for all folders.")
```

## generating cellpose_shape_features.csv

```python
import os
from glob import glob
import numpy as np
import pandas as pd
from skimage.io import imread
from skimage.measure import regionprops_table
from tqdm import tqdm

# Define paths
mask_base = '/content/cellpose_outputs'
image_base = '/content/drive/MyDrive/ProjectMIA/FinalFolder/3. HEtiles_small'
output_csv = '/content/cellpose_shape_features.csv'

all_data = []

# Loop over all folders
folders = sorted([f for f in os.listdir(mask_base) if os.path.isdir(os.path.join(mask_base, f))])
print(f"üìÇ Found {len(folders)} folders.")

for folder in folders:
    mask_folder = os.path.join(mask_base, folder)
    image_folder = os.path.join(image_base, folder)

    mask_paths = sorted(glob(os.path.join(mask_folder, '*_cp_masks.png')))
    print(f"\nüìÅ Processing {folder} with {len(mask_paths)} masks...")

    for mask_path in tqdm(mask_paths, desc=f"Analyzing {folder}"):
        base_name = os.path.basename(mask_path).replace('_cp_masks.png', '')
        image_path = os.path.join(image_folder, base_name + '.png')

        if not os.path.exists(image_path):
            print(f"‚ùå Image not found for: {base_name}")
            continue

        mask = imread(mask_path)
        img = imread(image_path)

        try:
            props = regionprops_table(
                mask,
                intensity_image=img,
                properties=[
                    'label', 'area', 'perimeter', 'eccentricity', 'solidity', 'extent',
                    'major_axis_length', 'minor_axis_length',
                    'mean_intensity', 'max_intensity', 'min_intensity',
                    'centroid'
                ]
            )
            df = pd.DataFrame(props)
            df['image_id'] = base_name
            df['folder'] = folder
            all_data.append(df)
        except Exception as e:
            print(f"‚ö†Ô∏è Error processing {base_name}: {e}")

# Combine and save CSV
feature_df = pd.concat(all_data, ignore_index=True)
feature_df.to_csv(output_csv, index=False)

print(f"\n‚úÖ Feature extraction complete! CSV saved at: {output_csv}")
```

## generating cellpose_newfeatures_combined.csv

```python
import os
import numpy as np
import pandas as pd
from skimage.io import imread
from skimage.measure import regionprops, label, perimeter
from tqdm import tqdm

# Define input and output paths
base_mask_path = "/content/cellpose_outputs"
output_csv = "/content/cellpose_newfeatures_combined.csv"

# Prepare to collect all results
all_features = []

# Loop through folders (folder1 to folder33)
for folder in sorted(os.listdir(base_mask_path)):
    folder_path = os.path.join(base_mask_path, folder)
    if not os.path.isdir(folder_path):
        continue

    # Loop through all mask images in the folder
    mask_files = sorted([f for f in os.listdir(folder_path) if f.endswith("_cp_masks.png")])
    for file in tqdm(mask_files, desc=f"Processing {folder}"):
        mask_path = os.path.join(folder_path, file)
        mask = imread(mask_path)

        # Label and analyze regions
        labeled = label(mask)
        props = regionprops(labeled)

        for region in props:
            area = region.area
            perim = perimeter(region.image, neighborhood=8)
            solidity = region.solidity
            major_axis = region.major_axis_length
            minor_axis = region.minor_axis_length
            centroid_x, centroid_y = region.centroid
            circularity = (4 * np.pi * area) / (perim ** 2) if perim != 0 else 0

            all_features.append({
                "Image": file.replace("_cp_masks.png", ""),
                "Folder": folder,
                "Centroid X px": centroid_x,
                "Centroid Y px": centroid_y,
                "Area px^2": area,
                "Length px": perim,
                "Circularity": circularity,
                "Solidity": solidity,
                "Max diameter px": major_axis,
                "Min diameter px": minor_axis
            })

# Convert to DataFrame and save
df = pd.DataFrame(all_features)
df.to_csv(output_csv, index=False)
print("‚úÖ Features saved to:", output_csv)
```

## generating cellpose_HE_features.csv

```python
import os
import cv2
import numpy as np
import pandas as pd
from tqdm import tqdm

# Path to H&E-stained images organized as folder1 to folder33
he_base_path = "/content/drive/MyDrive/ProjectMIA/FinalFolder/3. HEtiles_small"
output_csv = "/content/cellpose_HE_features.csv"

results = []

# Loop over all folders
for folder in sorted(os.listdir(he_base_path)):
    folder_path = os.path.join(he_base_path, folder)
    if not os.path.isdir(folder_path):
        continue

    print(f"üìÅ Processing {folder}...")
    image_files = sorted([f for f in os.listdir(folder_path) if f.endswith('.png')])

    for image_file in tqdm(image_files, desc=f"Processing {folder}", leave=False):
        image_path = os.path.join(folder_path, image_file)
        img = cv2.imread(image_path)
        if img is None:
            print(f"‚ö†Ô∏è Could not load image: {image_path}")
            continue

        # Convert to RGB from BGR (OpenCV loads as BGR)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Approximate channels:
        eosin = img_rgb[:, :, 0]     # Red channel ‚âà Eosin
        green = img_rgb[:, :, 1]     # Green is unused here
        hematoxylin = img_rgb[:, :, 2]  # Blue channel ‚âà Hematoxylin

        def extract_stats(channel):
            return {
                'Mean': np.mean(channel),
                'Median': np.median(channel),
                'Min': np.min(channel),
                'Max': np.max(channel),
                'Std': np.std(channel)
            }

        h_stats = extract_stats(hematoxylin)
        e_stats = extract_stats(eosin)

        results.append({
            'folder': folder,
            'image': image_file,
            # Hematoxylin (Blue)
            'Hematoxylin: Mean': h_stats['Mean'],
            'Hematoxylin: Median': h_stats['Median'],
            'Hematoxylin: Min': h_stats['Min'],
            'Hematoxylin: Max': h_stats['Max'],
            'Hematoxylin: Std': h_stats['Std'],
            # Eosin (Red)
            'Eosin: Mean': e_stats['Mean'],
            'Eosin: Median': e_stats['Median'],
            'Eosin: Min': e_stats['Min'],
            'Eosin: Max': e_stats['Max'],
            'Eosin: Std': e_stats['Std']
        })

# Save to CSV
df = pd.DataFrame(results)
df.to_csv(output_csv, index=False)
print(f"\n‚úÖ H&E feature extraction complete. CSV saved at: {output_csv}")
```

## generating cellpose_nuclear_features.csv

```python
import os
import numpy as np
import pandas as pd
from skimage.io import imread
from skimage.measure import regionprops, label
from skimage.draw import polygon
from skimage.color import rgb2hed
from skimage.filters import threshold_otsu
from tqdm import tqdm

# üìÅ Define input paths
image_base = "/content/drive/MyDrive/ProjectMIA/FinalFolder/3. HEtiles_small"
mask_base = "/content/cellpose_outputs"
output_csv = "/content/cellpose_nuclear_features.csv"

# üì¶ Helper function to compute circularity
def circularity(area, perimeter):
    if perimeter == 0:
        return 0
    return (4 * np.pi * area) / (perimeter ** 2)

# üß† Process all folders and images
all_data = []

folders = sorted([f for f in os.listdir(image_base) if os.path.isdir(os.path.join(image_base, f))])

for folder in tqdm(folders, desc="üìÇ Processing folders"):
    image_folder = os.path.join(image_base, folder)
    mask_folder = os.path.join(mask_base, folder)

    image_files = sorted([f for f in os.listdir(image_folder) if f.endswith(".png")])

    for img_file in image_files:
        img_path = os.path.join(image_folder, img_file)
        mask_file = img_file.replace(".png", "_cp_masks.png")
        mask_path = os.path.join(mask_folder, mask_file)

        if not os.path.exists(mask_path):
            print(f"‚ùå Missing mask: {mask_file}")
            continue

        # Load image and mask
        image = imread(img_path)
        mask = imread(mask_path)

        # Convert RGB to HED
        hed = rgb2hed(image)
        hematoxylin = hed[:, :, 0]
        eosin = hed[:, :, 1]

        labeled_mask = label(mask)
        props = regionprops(labeled_mask)

        for prop in props:
            coords = prop.coords
            rr, cc = coords[:, 0], coords[:, 1]

            area = prop.area
            perimeter = prop.perimeter
            circ = circularity(area, perimeter)
            solidity = prop.solidity
            major = prop.major_axis_length
            minor = prop.minor_axis_length
            centroid_y, centroid_x = prop.centroid  # (row, col)

            # Extract color intensities from H and E channels
            h_vals = hematoxylin[rr, cc]
            e_vals = eosin[rr, cc]

            data = {
                "Image": img_file,
                "Centroid X px": centroid_x,
                "Centroid Y px": centroid_y,
                "Area px^2": area,
                "Perimeter px": perimeter,
                "Circularity": circ,
                "Solidity": solidity,
                "Max diameter px": major,
                "Min diameter px": minor,
                "Hematoxylin: Mean": np.mean(h_vals),
                "Hematoxylin: Median": np.median(h_vals),
                "Hematoxylin: Min": np.min(h_vals),
                "Hematoxylin: Max": np.max(h_vals),
                "Hematoxylin: Std.Dev.": np.std(h_vals),
                "Eosin: Mean": np.mean(e_vals),
                "Eosin: Median": np.median(e_vals),
                "Eosin: Min": np.min(e_vals),
                "Eosin: Max": np.max(e_vals),
                "Eosin: Std.Dev.": np.std(e_vals),
                "Folder": folder
            }
            all_data.append(data)

# üíæ Save to CSV
df = pd.DataFrame(all_data)
df.to_csv(output_csv, index=False)
print(f"‚úÖ CSV saved to: {output_csv}")
```

## generating Eosin intensity vales per nucleus

```python
import os
import numpy as np
import pandas as pd
from skimage.io import imread
from skimage.measure import regionprops, label
from tqdm import tqdm

# Paths
he_base = "/content/drive/MyDrive/ProjectMIA/FinalFolder/3. HEtiles_small"
mask_base = "/content/cellpose_outputs"

results = []

# Loop through 33 folders
folders = sorted([f for f in os.listdir(mask_base) if os.path.isdir(os.path.join(mask_base, f))])
print(f"üìÅ Found {len(folders)} folders.")

for folder in folders:
    mask_folder = os.path.join(mask_base, folder)
    he_folder = os.path.join(he_base, folder)
    mask_files = sorted([f for f in os.listdir(mask_folder) if f.endswith("_cp_masks.png")])

    print(f"\nüîç Processing {folder} with {len(mask_files)} images")
    for mask_file in tqdm(mask_files, desc=f"Eosin stats in {folder}"):
        mask_path = os.path.join(mask_folder, mask_file)
        img_name = mask_file.replace("_cp_masks.png", ".png")
        img_path = os.path.join(he_folder, img_name)

        mask = imread(mask_path)
        img = imread(img_path)

        if mask.shape[:2] != img.shape[:2]:
            print(f"‚ö†Ô∏è Size mismatch: {img_name}")
            continue

        red_channel = img[:, :, 0]
        labeled = label(mask)

        for region in regionprops(labeled, intensity_image=red_channel):
            intensities = region.intensity_image[region.image]
            eosin_stats = {
                'folder': folder,
                'image': img_name,
                'nucleus_id': region.label,
                'eosin_mean': np.mean(intensities),
                'eosin_median': np.median(intensities),
                'eosin_min': np.min(intensities),
                'eosin_max': np.max(intensities),
                'eosin_std': np.std(intensities),
            }
            results.append(eosin_stats)

# Save to CSV
df = pd.DataFrame(results)
csv_path = "/content/drive/MyDrive/ProjectMIA/FinalFolder/cellpose_eosin_per_nucleus.csv"
df.to_csv(csv_path, index=False)

print(f"\n‚úÖ Saved Eosin stats per nucleus to: {csv_path}")
```

üìÅ Converts them to .xlsx

üìä Generates a summary file with nuclei counts

```python
import os
import pandas as pd
from tqdm import tqdm
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Paths
root_path = '/content/drive/My Drive/ProjectMIA/FinalFolder/4. QUPathCSVs/'
folder_names = [f'folder{i}' for i in range(1, 34)]

# Initialize
nuclei_summary = []
combined_df = pd.DataFrame()

# Set up progress bar for 165 files
total_files = 33 * 5
progress = tqdm(total=total_files, desc="Processing Files", unit="file")

# Loop through folders
for folder_name in folder_names:
    folder_path = os.path.join(root_path, folder_name)

    try:
        txt_files = sorted([f for f in os.listdir(folder_path) if f.endswith('.txt')])
    except FileNotFoundError:
        print(f"‚ùå Missing folder: {folder_path}")
        continue

    for idx, txt_file in enumerate(txt_files, 1):
        # Rename file
        new_name = f"{folder_name}_image{idx}"
        old_txt = os.path.join(folder_path, txt_file)
        new_txt = os.path.join(folder_path, f"{new_name}.txt")
        new_xlsx = os.path.join(folder_path, f"{new_name}.xlsx")
        os.rename(old_txt, new_txt)

        # Load, convert, and save
        df = pd.read_csv(new_txt, sep='\t')
        df.to_excel(new_xlsx, index=False)

        # Add to combined data
        df.insert(0, 'Image_ID', new_name)
        combined_df = pd.concat([combined_df, df], ignore_index=True)

        # Count nuclei
        nuclei_summary.append({'Image_ID': new_name, 'Nuclei_Count': len(df)})

        # Update progress
        progress.update(1)

# Final save
pd.DataFrame(nuclei_summary).to_excel(os.path.join(root_path, 'Nuclei_Count_Summary.xlsx'), index=False)
combined_df.to_excel(os.path.join(root_path, 'All_Images_Combined.xlsx'), index=False)

progress.close()
print("‚úÖ All files processed.")
```

## try out Cohen‚Äôs d

Cohen‚Äôs d is used to measure effect size (how strongly two groups differ), and it's ideal when:

- have two independent groups (like Cellpose vs StarDist)

- comparing a continuous feature

```python
import numpy as np

def cohens_d(a, b):
    """Calculate Cohen's d for two independent samples"""
    mean_diff = np.mean(a) - np.mean(b)
    pooled_std = np.sqrt((np.std(a, ddof=1) ** 2 + np.std(b, ddof=1) ** 2) / 2)
    return mean_diff / pooled_std

# Example: Cohen‚Äôs d for Solidity
cell_vals = cellpose_features['Solidity'].dropna()
star_vals = stardist_features['Solidity'].dropna()
d = cohens_d(cell_vals, star_vals)

print(f"Cohen‚Äôs d for Solidity: {d:.3f}")
```

 ## clustering performance in an unsupervised setting

(No Ground Truth Needed)
will compute:

1. Silhouette Score: How well each sample fits within its cluster.

2. Davies-Bouldin Index: Lower is better (compact and well-separated clusters).

3. Calinski-Harabasz Index: Higher is better (compact within-cluster and distinct between-cluster dispersion).

| Metric                   | Interpretation                  |
| ------------------------ | ------------------------------- |
| **Silhouette Score**     | Closer to 1 = clearer clusters  |
| **Davies-Bouldin Index** | Closer to 0 = better separation |
| **Calinski-Harabasz**    | Higher = dense, well-separated  |


```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.preprocessing import StandardScaler

# Load data
cellpose_path = '/content/drive/My Drive/ProjectMIA/FinalFolder/cellpose_nuclear_features.csv'
stardist_path = '/content/drive/My Drive/ProjectMIA/FinalFolder/4. QUPathCSVs/StarDist_nuclear_features.xlsx'

cellpose_df = pd.read_csv(cellpose_path)
stardist_df = pd.read_excel(stardist_path)

# Select common numeric features
common_features = ['Area px^2', 'Circularity', 'Solidity', 'Max diameter px', 'Min diameter px']

# Scale and evaluate for Cellpose
X_cellpose = cellpose_df[common_features].dropna()
X_cellpose_scaled = StandardScaler().fit_transform(X_cellpose)
kmeans_c = KMeans(n_clusters=3, random_state=42).fit(X_cellpose_scaled)
labels_c = kmeans_c.labels_

sil_c = silhouette_score(X_cellpose_scaled, labels_c)
db_c = davies_bouldin_score(X_cellpose_scaled, labels_c)
ch_c = calinski_harabasz_score(X_cellpose_scaled, labels_c)

# Scale and evaluate for StarDist
X_stardist = stardist_df[common_features].dropna()
X_stardist_scaled = StandardScaler().fit_transform(X_stardist)
kmeans_s = KMeans(n_clusters=3, random_state=42).fit(X_stardist_scaled)
labels_s = kmeans_s.labels_

sil_s = silhouette_score(X_stardist_scaled, labels_s)
db_s = davies_bouldin_score(X_stardist_scaled, labels_s)
ch_s = calinski_harabasz_score(X_stardist_scaled, labels_s)

# Print results
print("\nüîç Clustering Evaluation (KMeans, k=3):")
print("üìä Cellpose:")
print(f"   Silhouette Score        : {sil_c:.4f}")
print(f"   Davies-Bouldin Index    : {db_c:.4f}")
print(f"   Calinski-Harabasz Index : {ch_c:.4f}")

print("\nüìä StarDist:")
print(f"   Silhouette Score        : {sil_s:.4f}")
print(f"   Davies-Bouldin Index    : {db_s:.4f}")
print(f"   Calinski-Harabasz Index : {ch_s:.4f}")
```

What This Means:
StarDist shows tighter, more well-separated clusters. The model seems to capture biologically distinct nuclear phenotypes better.

Cellpose clusters are less distinct, possibly due to variability in nuclear boundary segmentation or over-segmentation.

```python
import pandas as pd
from scipy.stats import ttest_ind, mannwhitneyu
import numpy as np

# Load data
cellpose_df = pd.read_csv('/content/drive/My Drive/ProjectMIA/FinalFolder/cellpose_nuclear_features.csv')
stardist_df = pd.read_excel('/content/drive/My Drive/ProjectMIA/FinalFolder/4. QUPathCSVs/StarDist_nuclear_features.xlsx')

# Clean and align columns
cellpose_df.rename(columns={'Image': 'Image_ID'}, inplace=True)
stardist_df.rename(columns={'Image': 'Image_ID'}, inplace=True)

# Shared features for comparison
common_features = [
    'Centroid X px', 'Centroid Y px', 'Area px^2', 'Circularity', 'Solidity',
    'Max diameter px', 'Min diameter px',
    'Hematoxylin: Mean', 'Hematoxylin: Median', 'Hematoxylin: Min', 'Hematoxylin: Max', 'Hematoxylin: Std.Dev.',
    'Eosin: Mean', 'Eosin: Median', 'Eosin: Min', 'Eosin: Max', 'Eosin: Std.Dev.'
]

# Initialize results
ttest_pvals = []
mw_pvals = []
cohen_ds = []

# Run tests
for feat in common_features:
    c_values = cellpose_df[feat].dropna()
    s_values = stardist_df[feat].dropna()

    # t-test and Mann-Whitney U
    t_pval = ttest_ind(c_values, s_values, equal_var=False).pvalue
    mw_pval = mannwhitneyu(c_values, s_values, alternative='two-sided').pvalue

    # Cohen's d
    mean_diff = np.mean(c_values) - np.mean(s_values)
    pooled_std = np.sqrt(((len(c_values) - 1) * np.var(c_values, ddof=1) + (len(s_values) - 1) * np.var(s_values, ddof=1)) / (len(c_values) + len(s_values) - 2))
    cohen_d = mean_diff / pooled_std

    ttest_pvals.append(t_pval)
    mw_pvals.append(mw_pval)
    cohen_ds.append(cohen_d)

# Create summary DataFrames
pvals_df = pd.DataFrame({
    'Feature': common_features,
    't-test p-value': ttest_pvals,
    'Mann-Whitney U p-value': mw_pvals
})

cohend_df = pd.DataFrame({
    'Feature': common_features,
    'Cohen_d': cohen_ds
})

# Now calculate the mean stats again
mean_pval_c = pvals_df['t-test p-value'].mean()
mean_mw_pval_c = pvals_df['Mann-Whitney U p-value'].mean()
mean_cohen_d = cohend_df['Cohen_d'].mean()

print("Mean t-test p-value:", mean_pval_c)
print("Mean Mann-Whitney U p-value:", mean_mw_pval_c)
print("Mean Cohen's d:", mean_cohen_d)
```

```python
import pandas as pd

# Metric values
metrics = [
    "Silhouette Score",
    "Davies-Bouldin Index",
    "Calinski-Harabasz Index",
    "Mean |Cohen‚Äôs d|",
    "Mean p-value (t-test)",
    "Total Nuclei Count"
]

cellpose_vals = [0.3753, 0.9597, 33062.5898, 2.11, 0.0413, 59979]
stardist_vals = [0.4352, 0.8033, 101863.2783, 2.31, 0.0357, 121775]

# For direction: +1 means higher is better, -1 means lower is better
directions = [1, -1, 1, 1, -1, 1]

# Weights assigned to each metric
weights = [0.2, 0.1, 0.2, 0.2, 0.1, 0.2]

# Explanation for each metric
explanations = [
    "Measures cluster separation; higher is better.",
    "Measures intra-cluster similarity; lower is better.",
    "Higher score means better-defined clusters.",
    "Effect size: how different the models are in feature distribution.",
    "Smaller p-values indicate stronger statistical differences.",
    "Models that detect more valid nuclei are preferred."
]

# Normalize metrics for scoring
def normalize(val1, val2, direction):
    if direction == 1:
        total = val1 + val2
        return val1 / total, val2 / total
    else:
        total = 1 / val1 + 1 / val2
        return (1 / val1) / total, (1 / val2) / total

# Normalize and compute scores
normalized_c = []
normalized_s = []
for i in range(len(metrics)):
    c, s = normalize(cellpose_vals[i], stardist_vals[i], directions[i])
    normalized_c.append(c)
    normalized_s.append(s)

# Final weighted scores
final_score_c = sum([c * w for c, w in zip(normalized_c, weights)])
final_score_s = sum([s * w for s, w in zip(normalized_s, weights)])

# Build table
df = pd.DataFrame({
    "Metric": metrics,
    "Explanation": explanations,
    "Cellpose": cellpose_vals,
    "StarDist": stardist_vals,
    "Weight": weights,
    "Direction": ["‚Üë" if d==1 else "‚Üì" for d in directions],
    "Normalized Cellpose": [round(val, 4) for val in normalized_c],
    "Normalized StarDist": [round(val, 4) for val in normalized_s]
})

# Append total row
df.loc["Total Score"] = [
    "", "Weighted sum of normalized scores",
    "", "", "", "",
    round(final_score_c, 4), round(final_score_s, 4)
]

# Show table
import IPython.display as display
display.display(df)
```

```python
import pandas as pd

# Raw values
metrics = {
    'Silhouette': {'Cellpose': 0.3753, 'StarDist': 0.4352},
    'DBI': {'Cellpose': 0.9597, 'StarDist': 0.8033},
    'CH': {'Cellpose': 33062.5898, 'StarDist': 101863.2783},
    'pval': {'Cellpose': 0.0865, 'StarDist': 0.0865},
    'cohen': {'Cellpose': 1.911, 'StarDist': 1.911},
    'nuclei': {'Cellpose': 59979, 'StarDist': 121775}
}

# Weighting
weights = {
    'Silhouette': 0.25,
    'DBI': 0.15,
    'CH': 0.20,
    'pval': 0.10,
    'cohen': 0.20,
    'nuclei': 0.10
}

# Normalize
def normalize(metric_dict, higher_is_better=True):
    vals = list(metric_dict.values())
    min_val, max_val = min(vals), max(vals)
    if max_val - min_val == 0:
        return {k: 0.5 for k in metric_dict}  # neutral score
    if higher_is_better:
        return {k: (v - min_val) / (max_val - min_val) for k, v in metric_dict.items()}
    else:
        return {k: (max_val - v) / (max_val - min_val) for k, v in metric_dict.items()}

# Normalize all
norm_metrics = {
    'Silhouette': normalize(metrics['Silhouette'], True),
    'DBI': normalize(metrics['DBI'], False),
    'CH': normalize(metrics['CH'], True),
    'pval': normalize(metrics['pval'], False),
    'cohen': normalize(metrics['cohen'], True),
    'nuclei': normalize(metrics['nuclei'], True)
}

# Compute weighted scores
final_scores = {'Cellpose': 0, 'StarDist': 0}
for metric, norm_values in norm_metrics.items():
    for model in final_scores:
        final_scores[model] += norm_values[model] * weights[metric]

# Output
print("üîç Weighted Scores:")
for model, score in final_scores.items():
    print(f"{model}: {score:.4f}")

winner = max(final_scores, key=final_scores.get)
print(f"\nüèÜ Best model based on weighted criteria: **{winner}**")
```
